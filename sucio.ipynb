```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.concat([zoo2, zoo3])
data
```

```python
data.drop_duplicates()
data.reset_index(drop=True)
data
```

```python
fig, ax = plt.subplots(5,3, figsize=(20,18))
sns.countplot(data['hair'], ax=ax[0,0])
sns.countplot(data['feathers'], ax=ax[0,1])
sns.countplot(data['eggs'], ax=ax[0,2])
sns.countplot(data['milk'], ax=ax[1,0])
sns.countplot(data['airborne'], ax=ax[1,1])
sns.countplot(data['predator'], ax=ax[1,2])
sns.countplot(data['toothed'], ax=ax[2,0])
sns.countplot(data['backbone'], ax=ax[2,1])
sns.countplot(data['breathes'], ax=ax[2,2])
sns.countplot(data['venomous'], ax=ax[3,0])
sns.countplot(data['fins'], ax=ax[3,1])
sns.countplot(data['legs'], ax=ax[3,2])
sns.countplot(data['tail'], ax=ax[4,0])
sns.countplot(data['catsize'], ax=ax[4,1])
sns.countplot(data['class_type'], ax=ax[4,2])
```
